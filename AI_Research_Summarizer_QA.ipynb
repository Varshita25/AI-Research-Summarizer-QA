## **Upload and Read the PDF**



```python
import pdfplumber

def extract_text_from_pdf(pdf_path):
    text = ""
    with pdfplumber.open(pdf_path) as pdf:
        for page in pdf.pages:
            text += page.extract_text()
    return text

# Upload PDF
from google.colab import files
uploaded = files.upload()

# Use the first uploaded file
file_path = list(uploaded.keys())[0]
pdf_text = extract_text_from_pdf(file_path)
print(pdf_text[:1000])  # preview
```

## **Summarize Using BART**

```python
from transformers import pipeline

# Load summarization pipeline
summarizer = pipeline("summarization", model="facebook/bart-large-cnn")

def summarize_text(text, max_chunk=500):
    summary = ""
    text = text.replace('\n', ' ')
    chunks = [text[i:i+max_chunk] for i in range(0, len(text), max_chunk)]

    for chunk in chunks:
        summary_piece = summarizer(chunk, max_length=130, min_length=30, do_sample=False)[0]['summary_text']
        summary += summary_piece + " "

    return summary

summary_output = summarize_text(pdf_text[:3000])  # limit to 3000 chars for free tier
print("🔍 Summary:\n", summary_output)
```

## **Ask Questions Using DistilBERT QA**

```python
from transformers import pipeline

qa_pipeline = pipeline("question-answering", model="distilbert-base-uncased-distilled-squad")

def ask_question(context, question):
    return qa_pipeline(question=question, context=context)['answer']

# Example
question = "What is the main contribution of the paper?"
answer = ask_question(pdf_text, question)
print("Q:", question)
print("A:", answer)
```

## **Create a Gradio UI**

```python
import gradio as gr

def summarizer_interface(pdf_file):
    with pdfplumber.open(pdf_file.name) as pdf:
        full_text = "".join([page.extract_text() for page in pdf.pages])

    summary = summarize_text(full_text[:3000])
    return summary

def qa_interface(pdf_file, question):
    with pdfplumber.open(pdf_file.name) as pdf:
        full_text = "".join([page.extract_text() for page in pdf.pages])

    answer = ask_question(full_text, question)
    return answer

summary_ui = gr.Interface(fn=summarizer_interface, inputs="file", outputs="text", title="Research Paper Summarizer")
qa_ui = gr.Interface(fn=qa_interface, inputs=["file", "text"], outputs="text", title="Research Paper Q&A Tool")

gr.TabbedInterface([summary_ui, qa_ui], ["Summarizer", "Q&A"]).launch()
```
